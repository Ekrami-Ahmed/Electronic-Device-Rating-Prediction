```python
has_ssd = []
for index, row in df.iterrows():
    ssd = row['ssd']
    hdd = row['hdd']
    
    if ssd == '0 GB' and hdd != '0 GB':
        has_ssd.append('has only hdd')
    elif ssd != '0 GB' and hdd == '0 GB':
        has_ssd.append('has only ssd')
    else:
        has_ssd.append('has ssd and hdd')

df['has_ssd'] = has_ssd
df.head()
```

## Encoding

```python
encode = ['brand','processor_brand','processor_name','ram_type','os','weight','warranty', 'has_ssd', 'processor_gnrtn']
one_hot = ['Touchscreen','msoffice']
s_encode = df[encode]
one_encode = df[one_hot]
df.processor_name.unique()
```

### Label encoding

```python
label_encoder = LabelEncoder()

for column in encode:
    s_encode[column] = label_encoder.fit_transform(s_encode[column])


s_encode.head()
```

### One-hot encoding

```python
one_encoded_data = pd.get_dummies(one_encode, columns=one_hot)
one_encoded_data [one_encoded_data  != False] = 1
one_encoded_data [one_encoded_data  != True] = 0
one_encoded_data.head()
```

```python
# Combine Label and One-hot encoding
result = pd.concat([s_encode, one_encoded_data], axis=1)
result.head()
```

```python
colToProcess=['ram_gb','ssd','hdd','graphic_card_gb','rating']
for col in colToProcess:
  splitted_data = df[col].str.split(' ')
  df[col] = [row[0] for row in splitted_data]
  df[col] = df[col].astype(int)
  
df.head()
```

```python
df.drop('Touchscreen', axis=1, inplace=True)
df.drop('msoffice', axis=1, inplace=True)
```

```python
df[result.columns] = result
df.head()
```
